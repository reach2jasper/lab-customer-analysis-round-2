### 02 - Getting Data


```python
import pandas as pd
import numpy as nm
market=pd.read_csv('marketing_customer_analysis.csv')
```

### 03 - Cleaning/Wrangling/EDA

```python
market.head()
```

```python
#Deleting unwanted columns (Unnamed: 0)
market = market.drop(['Unnamed: 0'], axis=1)
market.head()
```

```python
# Change headers names to lowercase
cols = []
for column in market.columns:
    cols.append(column.lower())
cols
market.columns = cols
market.head()
```

```python
# Removing Duplicates
print(market.shape)
market2 = market.copy()
market2 = market2.drop_duplicates()  
print(market2.shape)
```

```python
# Dealing with NaN values
market2.isna().sum()
market2['state'] = market2['state'].fillna('California') # Changed the value of Nan in 'state' to 'California'
market2['response'] = market2['response'].fillna('No') # Changed the value of Nan in 'response' to 'No'
market2=market2.dropna(axis=0)# Removed all the rows which has Nan
market2.shape
```

```python
# Categorical Features
market_cat=market2.select_dtypes('object')
market_cat.head()
```

```python
# Numerical Features
market_num=market2._get_numeric_data()
market_num.head()
```

```python
# Exploration
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
sns.barplot(x="gender", y="total claim amount", data=market2)
plt.show()
```

```python
sns.boxplot(x = 'gender',y="total claim amount", data=market2)
plt.show()
```

```python
market['total claim amount'].hist(bins=100)
plt.show()
```

```python
#Correlation Matrix
correlations_matrix = market2.corr()
correlations_matrix
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```
