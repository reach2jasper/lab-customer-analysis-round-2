```python
# BONUS: Put all the previously mentioned data transformations into a function.
def market_analysis():
    mark=pd.read_csv('marketing_customer_analysis.csv')
    mark=mark.drop('Unnamed: 0', axis=1)
    mark.shape
    cols=[]
    for column in mark.columns:
        cols.append(column.lower())
    cols
    mark.columns=cols
    mark
    display(mark.head(0))
    mark._get_numeric_data()
    mark.select_dtypes('object')
    mark.isna().sum()
    mark = mark[mark['state'].isna()==False] # Since these nulls are not a lot, we can filter them
    mark.isna().sum()
    mark.shape
    mark['effective to date'] = pd.to_datetime(mark['effective to date'], errors='coerce')
    mark.dtypes
    mark['month']=mark['effective to date'][0].month
    mark['month']
    mark[mark['month']<2]
    return mark
```

```python
#Normalize (numerical)
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
transformer = MinMaxScaler().fit(X_num)
x_normalized = transformer.transform(X_num)
print(x_normalized.shape)
x_normalized
pd.DataFrame(x_normalized, columns=X_num.columns)
```

# Lab | Customer Analysis Round 6

```python
# One Hot/Label Encoding (categorical). I have used Dummy variable encoding from Pandas
X_cat.head()
X_cat['gender'].unique()
X_catgender = pd.get_dummies(X_cat['gender'])
```

```python
# Concat DataFrames
X = nm.concatenate([x_normalized, X_catgender], axis=1)
```

```python
# Linear Regression
from sklearn import linear_model
lm = linear_model.LinearRegression()
X1=mark['income']
y1=mark['total claim amount']
lm.fit(X1,y1)
```
